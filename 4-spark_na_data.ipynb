```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('na').getOrCreate()
```

```
df = spark.read.csv("dbfs:/FileStore/tables/ContainsNull.csv", inferSchema=True,
                                                               header=True)
```

```
df.show()
```

```
df.na.drop(thresh=2).show() 
```

```
df.na.drop(subset=['Sales']).show()
```

```
df.printSchema()
```

```
df.na.fill('FILL VALUE').show() # Spark only fills cells that have the same data type as specificed in "fill" function. In this case na rows in Sales columns are not filled since they are expected to be doubles.
```

```
df.na.fill('No Name', subset=['Name']).show()
```

```
from pyspark.sql.functions import mean
```

```
mean_val = df.select(mean(df['Sales'])).collect()[0][0]
```

```
df.na.fill(mean_val,['Sales']).show()
```

```
df.na.fill(df.select(mean(df['Sales'])).collect()[0][0],['Sales']).show()
```
